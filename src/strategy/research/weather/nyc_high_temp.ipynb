```python
from datetime import datetime
def parse_datetime(datetime_str):
    # Specify the format of the input datetime string
    datetime_format = "%b %d, %I:%M %p, %Y"  # %b for abbreviated month name, %I for 12-hour hour, %M for minute, %p for AM/PM

    # Concatenate the provided year with the input datetime string
    datetime_str_with_year = f"{datetime_str}, 2024"

    # Parse the datetime string into a datetime object
    parsed_datetime = datetime.strptime(datetime_str_with_year, datetime_format)

    return parsed_datetime
```
